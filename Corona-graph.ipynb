```python
import pandas as pd 
import matplotlib as mpl
from pandas import DataFrame 
import matplotlib.pyplot as plt 
from matplotlib.pyplot import figure 

df = pd.read_csv('C:/Users/Park/Desktop/graph/Corona.csv', encoding='cp949') 
fig = mpl.pyplot.gcf() 
plt.rcParams["figure.figsize"] = (12,4)
plot = df.plot() 
plot.set_xlabel("date")

plt.axvline(x=55, color='b', linestyle='--', linewidth=3) #사회적 거리두기 시행
plt.axvline(x=30 , color='b', linestyle='--', linewidth=3) #드라이브 수로 선별진료소 첫 시행
plt.axvline(x=22 , color='r', linestyle='--', linewidth=3) #31번 확진자
plt.axvline(x=24 , color='r', linestyle='--', linewidth=3) #국내 코로나 첫 사망자 발생
plt.axvline(x=42 , color='b', linestyle='--', linewidth=3) # 마스크 5부제
plt.axvline(x=88 , color='b', linestyle='--', linewidth=3) #재난 지원금 발표



plot.set_ylabel("confirmer")
plt.title("The Number of confirmers") 
plt.xlabel("1.20 ~ 4.26 dates") 
plt.ylabel("confirmer")

plt.show() 
```
